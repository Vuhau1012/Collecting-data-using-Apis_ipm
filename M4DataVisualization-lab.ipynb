## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
import matplotlib.pyplot as plt
import seaborn as sns
QUERY = """
SELECT ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY,conn)
# Vẽ histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df_ConvertedComp['ConvertedComp'], kde=True)  # Thay 'Column_name' bằng tên cột chứa dữ liệu
plt.title('Histogram of Data')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
import scipy.stats as stats

# Vẽ đường cong phân phối
plt.figure(figsize=(8, 6))
stats.probplot(df_ConvertedComp['ConvertedComp'], plot=plt)
plt.title('Probability Plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.show()
```

```python
from scipy.stats import shapiro

# Kiểm định Shapiro-Wilk
stat, p = shapiro(df_ConvertedComp['ConvertedComp'])
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Dữ liệu có phân phối chuẩn (không thể bác bỏ giả thuyết H0)')
else:
    print('Dữ liệu không có phân phối chuẩn (có bằng chứng để bác bỏ giả thuyết H0)')
```

```python
from scipy.stats import kstest

# Kiểm định Kolmogorov-Smirnov
stat, p = kstest(df_ConvertedComp['ConvertedComp'], 'norm')
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Dữ liệu có phân phối chuẩn (không thể bác bỏ giả thuyết H0)')
else:
    print('Dữ liệu không có phân phối chuẩn (có bằng chứng để bác bỏ giả thuyết H0)')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
QUERY = """
SELECT ConvertedComp
FROM master
"""
df_ConvertedComp = pd.read_sql_query(QUERY, conn)

# Check for NaN values
print("Number of NaN values in 'ConvertedComp':", df_ConvertedComp['ConvertedComp'].isnull().sum())

# Drop rows with NaN values
df_ConvertedComp.dropna(inplace=True)

# Calculate quartiles
Q1 = np.percentile(df_ConvertedComp['ConvertedComp'], 25)
Q3 = np.percentile(df_ConvertedComp['ConvertedComp'], 75)

# Calculate interquartile range (IQR)
IQR = Q3 - Q1

# Determine outlier cutoff
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
ConvertedComp_no_outliers = df_ConvertedComp[(df_ConvertedComp['ConvertedComp'] >= lower_bound) & (df_ConvertedComp['ConvertedComp'] <= upper_bound)]

# Plot histogram
plt.hist(ConvertedComp_no_outliers, bins=10, edgecolor='black')  # Adjust the number of bins as needed

# Add labels and title
plt.title('Histogram of ConvertedComp')
plt.xlabel('Converted Compensation')
plt.ylabel('Frequency')

# Show plot
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY = """
SELECT Age, WorkWeekHrs, CodeRevHrs
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
# Xóa các dòng có giá trị NaN (missing values)
df = df.dropna()
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]
median_hours_by_age = filtered_df.groupby('Age')[['WorkWeekHrs', 'CodeRevHrs']].median().reset_index()
import matplotlib.pyplot as plt

# Set up the figure size
plt.figure(figsize=(10, 6))

# Plotting the stacked bar chart
plt.bar(median_hours_by_age['Age'], median_hours_by_age['WorkWeekHrs'], label='WorkWeekHrs')
plt.bar(median_hours_by_age['Age'], median_hours_by_age['CodeRevHrs'], bottom=median_hours_by_age['WorkWeekHrs'], label='CodeRevHrs')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median WorkWeekHrs and CodeRevHrs for Ages 30-35')
plt.legend()

# Show plot
plt.show()

```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import matplotlib.pyplot as plt
import numpy as np
QUERY = """
SELECT Age, ConvertedComp
FROM master
"""
df = pd.read_sql_query(QUERY,conn)
# Xóa các dòng có giá trị NaN (missing values)
df = df.dropna()
#Filter the data to include only ages from 45 to 60:
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]
median_comp_by_age = filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()
import matplotlib.pyplot as plt

# Set up the figure size
plt.figure(figsize=(10, 6))

# Plotting the median ConvertedComp
plt.plot(median_comp_by_age['Age'], median_comp_by_age['ConvertedComp'], marker='o', linestyle='-')

# Adding labels and title
plt.xlabel('Age')
plt.ylabel('Median Converted Compensation')
plt.title('Median Converted Compensation for Ages 45-60')
plt.grid(True)  # Adding grid for better readability

# Show plot
plt.show()

```
